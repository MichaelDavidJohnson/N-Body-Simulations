```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

```python
class Object:
    
    
    def __init__(self,mass,location,velocity):
        self.mass = mass
        self.location = location
        self.velocity = velocity
        self.storedVelocity = [0,0]
    
    def updateVelocity(self,velocity):
        self.velocity = velocity
    
    def updateLocation(self,location):
        self.location = location
    
    def setStoredVelocity(self,velocity):
        self.storedVelocity = velocity
    
    def getStoredVelocity(self):
        return self.storedVelocity
    
    def getMass(self):
        return self.mass

    def getLocation(self):
        return self.location
    
    def getVelocity(self):
        return self.velocity
```

```python
G = 6.67408e-11
```

```python
def update_two(A,B):
    Force_Direction = (A.getLocation()-B.getLocation())/np.linalg.norm(A.getLocation()-B.getLocation())
    F = Force_Direction*G*(A.getMass()*B.getMass()/np.linalg.norm(A.getLocation()-B.getLocation())**2)
    A_Acc = F/A.getMass()
    B_Acc = F/B.getMass()
    A_Update = A.getVelocity()+A_Acc*t
    B_Update = B.getVelocity()+B_Acc*t
    Avelocity = A_Update
    Bvelocity = B_Update
    A.setStoredVelocity(A.getStoredVelocity() + Avelocity)
    B.setStoredVelocity(B.getStoredVelocity() + Bvelocity)
```

```python
x1 = np.array([0.97000436,-0.24308753])
x2 = -x1
x3 = 0
Vx3 = np.array([0.93240737, - 0.86473146])
Vx1 = -Vx3/2
Vx2 = -Vx3/2
```

```python
objectA = Object(1,x1,Vx1)
objectB = Object(1,x2,Vx2)
objectC = Object(1,x3,Vx3)
timestep = np.linspace(0,1,10000)
objectArray = [objectA,objectB,objectC]
```

```python
for t in timestep:
    for objects in objectArray:
        objects.setStoredVelocity(np.array([0,0]))
    pairs = itertools.combinations(objectArray,2) #This is clearly inefficient. This exists for futureproofing when popping elements out the array is discussed.
    
    for pair in pairs:
        update_two(pair[0],pair[1])
        update_two(pair[1],pair[0])
        
    
    for objects in objectArray:
        objects.updateVelocity(objects.getStoredVelocity())
        objects.updateLocation(objects.getLocation() + objects.getVelocity()) #Forward Euler
        print(objects.getLocation())
```
